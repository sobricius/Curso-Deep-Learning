# PRÁCTICA DE SEABORN CON PANDAS

### 0.- Importa Seaborn y Pandas

```python
import seaborn as sns
import pandas as pd
import numpy as np                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                               
```

### 1.- Carga del dataset
Usaremos el dataset *penguins* proporcionado por la propia librería *seaborn*.
La carga se realiza con el método `load_dataset()`.
Mostrar el número de filas del dataset y el nombre de cada locumna.
Mostrar también las 5 primeras filas.

```python
df=sns.load_dataset('penguins')
print(df.shape[0])
columns=df.columns.tolist();
print(columns)
print(df.head(5))
```

### 3.- Usar una gráfica *countplot* para mostrar el número de individuos de cada especie

```python
sns.countplot(hue='species',x='species',data=df)
```

### 4.- Usar una gráfica *displot* para mostrar la distribución de la masa corporal (*body_mass_g*). Comprobar el efecto del parámetro `kind='kde'`.

```python
sns.displot(data=df['body_mass_g'],kind='kde')
```

### 5.- Modificar el ejercicio anterior para mostrar simultáneamente la distribución para cada una de las especies.

```python
sns.displot(data=df,x='body_mass_g' ,hue='species',kind='kde')
```

### 6.- Mostrar un diagrama de cajas para la masa corporal para cada especie.

```python
sns.boxplot(df,y='body_mass_g',x='species',hue='species')
```

### 7.- Usar una gráfica *barplot* para mostrar la distribución de la longitud del pico (*bill_length_mm*) por islas.

```python
sns.barplot(data=df,hue='island',x='island',y='bill_length_mm')
```

### 8.- Usar *pairplot* para mostrar la relación entre las columnas numéricas del dataframe diferenciando por especie (columna *species*)

```python
sns.pairplot(df,hue='species')
```

### 9.- Usar un mapa de calor (*heatmap*) para mostrar la correlación entre las distintas variables numéricas del *dataframe*. Usar el método `corr()` de Pandas para el cálculo de las correlaciones.


```python
df.dtypes
numeric_df=df.select_dtypes(np.number)
correlacion_df=numeric_df.corr()
sns.heatmap(correlacion_df)
```
