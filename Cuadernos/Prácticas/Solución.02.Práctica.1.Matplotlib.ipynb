### 5.- Leer y mostrar el beneficio total (total_ventas) de todos los meses usando un gráfico de línea.

```
totalVentas = df ['total_ventas'].tolist()
meses = df ['num_mes'].tolist()
plt.plot(meses, totalVentas, label = 'Beneficio por mes')
plt.xlabel('Número de mes')
plt.ylabel('Ventas')
plt.xticks(meses)
plt.title('Ventas por mes')
plt.yticks([100000, 200000, 300000, 400000, 500000]);
```

### 6.-	Mostrar el beneficio total de todos los meses con un gráfico de línea (como en el ejercicio anterior) pero con las siguientes características de estilo:
- Línea roja discontinua de ancho 3.
- Leyenda en la posición inferior derecha.
- Leyenda del eje X: Número de mes
- Leyenda del eje Y: Número de unidades vendidas
- Añadir un marcador circular con línea roja.




```
plt.plot(meses, totalVentas, label = 'Ventas año pasado',
         color='r', marker='o', markerfacecolor='k',
         linestyle='--', linewidth=3)

plt.xlabel('Número de mes')
plt.ylabel('Beneficio')
plt.legend(loc='lower right')
plt.title('Ventas por mes')
plt.xticks(meses)
plt.yticks([100000, 200000, 300000, 400000, 500000]);
```

### 7.- Mostrar el número de unidades vendidas por mes de cada producto usando un gráfico multilínea (un gráfico de línea para cada producto). Mostrar una leyenda en la esquina superior izquierda con los colores de cada producto.

```
ventasCremaFacial      = df ['crema_facial'].tolist()
ventasLimpiadorFacial  = df ['limpiador_facial'].tolist()
ventasCremaDentifrica  = df ['crema_dentifrica'].tolist()
ventasGelBaño          = df ['gel_baño'].tolist()
ventasChampu           = df ['champu'].tolist()
ventasCremaHidratante  = df ['crema_hidratante'].tolist()

plt.plot(meses, ventasCremaFacial,       label = 'Crema facial',      marker='o', linewidth=3);
plt.plot(meses, ventasLimpiadorFacial,   label = 'Limpiador facial',  marker='o', linewidth=3)
plt.plot(meses, ventasCremaDentifrica,   label = 'Dentífrico',        marker='o', linewidth=3)
plt.plot(meses, ventasGelBaño,           label = 'Gel de baño',       marker='o', linewidth=3)
plt.plot(meses, ventasChampu,            label = 'Champú',            marker='o', linewidth=3)
plt.plot(meses, ventasCremaHidratante,   label = 'Crema hidratante',  marker='o', linewidth=3)

plt.xlabel('Número de mes')
plt.ylabel('Unidades')
plt.legend(loc='upper left')
plt.xticks(meses)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Información de ventas');
```

### 8.- Leer las ventas de pasta dentífrica de cada mes y mostrarlas usando un gráfico de dispersión (*scatter plot*). Además, se mostrará el gráfico con  rejilla (*grid*).

```
plt.scatter(meses, ventasCremaDentifrica, label = 'Ventas de dentífricos')
plt.xlabel('Número de mes')
plt.ylabel('Unidades vendidas')
plt.legend(loc='upper left')
plt.title('Ventas de dentífricos ')
plt.xticks(meses)
plt.grid(True, linewidth= 1, linestyle="--")
```

### 9.- Leer las ventas de crema facial y de crema limpiadora facial y mostrarlas en un gráfico de barras. El gráfico debe mostrar el número de unidades vendidas de cada producto por mes. Se debe mostrar una barra diferente para cada uno de esos dos productos.

```
# lista con las posiciones de las barras de crema facial:
posCremaFacial = []
for a in meses:
  posCremaFacial.append(a-0.25)
# posCremaFacial = [a-0.25 for a in meses] # alternativa más breve con list comprehension

posLimpiadorFacial = [a+0.25 for a in meses]

plt.bar(posCremaFacial, ventasCremaFacial, width= 0.25, label = 'Crema facial', align='edge')
plt.bar(posLimpiadorFacial, ventasLimpiadorFacial, width= -0.25, label = 'Limpiador facial', align='edge')
plt.xlabel('Número de mes')
plt.ylabel('Unidades vendidas')
plt.legend(loc='upper left')
plt.title('Información de ventas')
plt.xticks(meses)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title(' Ventas de crema facial y de limpiador facial');
```

### 10.-	Leer las ventas de gel de baño de todos los meses y mostrarlo con un gráfico de barras. Guardar el gráfico en fichero (en Google Drive).

### 11.- Leer las ventas totales de cada mes y mostrarlo usando un histograma. Los intervalos del histograma (*bins*) serán de 25000 € y el rango completo va de 150000 300000.

```
rangoVentas = list(range(150000, 300000, 25000))
plt.hist(totalVentas, rangoVentas, label = 'Ventas mensuales' ,rwidth = 0.9)
plt.xlabel('Rango de ventas en euros')
plt.ylabel('Número de meses')
plt.legend(loc='upper right')
plt.xticks(rangoVentas)
plt.title('Histograma de datos de ventas mensuales');
```

### 12.- Calcular las ventas anuales totales de cada producto y mostrarlo en un gráfico de tarta. Mostrar en cada sector el porcentaje que representa su producto sobre el total.

```
etiquetas = ['Crema facial', 'Limpiador facial', 'Crema dentífrica',
          'Gel de baño', 'Champú', 'Crema hidratante']

ventasPorProducto = [sum(ventasCremaFacial), sum(ventasLimpiadorFacial),
                     sum(ventasCremaDentifrica), sum(ventasGelBaño),
                     sum(ventasChampu), sum(ventasCremaHidratante)]
plt.figure(figsize=(9, 6))
plt.pie(ventasPorProducto, labels=etiquetas, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Ventas por producto');
```

### 13.- Leer las ventas de todos los meses de gel de baño y de limpiador facial y mostrarlos como dos subgráficos (subplots).

```
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(meses, ventasGelBaño, color='c', marker='o', linewidth=3)
axarr[0].set_title('Ventas de gel de baño')
axarr[1].plot(meses, ventasLimpiadorFacial, color='r', marker='o', linewidth=3)
axarr[1].set_title('Ventas de limpiador facial')

plt.xticks(meses)
plt.xlabel('Número de mes')
plt.ylabel('Unidades vendidas');
```

### 14.- Leer las ventas de todos los productos y mostrarlos con un gráfico de área (*stack plot*).

```
plt.figure(figsize=(8, 6))
plt.plot([],[],color='m', label='Crema facial', linewidth=5)
plt.plot([],[],color='y', label='Limpiador facial', linewidth=5)
plt.plot([],[],color='r', label='Crema dentífrica', linewidth=5)
plt.plot([],[],color='c', label='Gel de baño', linewidth=5)
plt.plot([],[],color='g', label='Champú', linewidth=5)
plt.plot([],[],color='b', label='Crema hidratante', linewidth=5)

plt.stackplot(meses, ventasCremaFacial, ventasLimpiadorFacial, ventasCremaDentifrica,
              ventasGelBaño, ventasChampu, ventasCremaHidratante,
              colors=['m','y','r','c','g','b'])

plt.xlabel('Número de mes')
plt.ylabel('Unidades vendidas')
plt.title('Datos de ventas de todos los productos')
plt.legend(loc='upper left');
```
