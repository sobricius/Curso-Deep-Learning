### 1.- Importar Pandas y Matplotlib

```python
import matplotlib.pyplot  as plt
import pandas as pd
import numpy as np
%matplotlib inline

```

### 3.- Carga del dataset 'ventas.csv'. Mostrar el nombre de las columnas y el contenido de las cinco primeras filas.

```python
df=pd.read_csv("/home/sergio/Documents/Cursos/Curso-Deep-Learning/Cuadernos/Prácticas/02.2.ventas.csv")
print(df.columns)
df.head(5)

```

### 4.- Establecer el tamaño por defecto de las figuras en 9.6 (ancho) x 7.2 (alto)

```python
plt.rcParams['figure.figsize']=[6,4.5]
```

### 5.- Leer y mostrar el beneficio total (total_ventas) de todos los meses usando un gráfico de línea.

```python
beneficio_total=df['total_ventas'].tolist()
meses=df['num_mes'].tolist()
plt.plot(meses,beneficio_total)
plt.xlabel('Numero de mes')
plt.ylabel('Ventas')
plt.xticks(meses)
plt.title('Ventas mensuales')
plt.yticks(np.arange(100000,600000,100000));

```

### 6.-	Mostrar el beneficio total de todos los meses con un gráfico de línea (como en el ejercicio anterior) pero con las siguientes características de estilo:
- Línea roja discontinua de ancho 3.
- Leyenda en la posición inferior derecha.
- Leyenda del eje X: Número de mes
- Leyenda del eje Y: Número de unidades vendidas
- Añadir un marcador circular con línea roja.




```python
beneficio_total=df['total_ventas'].tolist()
meses=df['num_mes'].tolist()
plt.plot(meses,beneficio_total,color='red',linestyle='--',linewidth=2,marker='o',markerfacecolor='k')
plt.xlabel('Numero de mes')
plt.ylabel('Ventas')
plt.xticks(meses)
plt.title('Ventas mensuales')
plt.yticks(np.arange(100000,600000,100000));

```

### 7.- Mostrar el número de unidades vendidas por mes de cada producto usando un gráfico multilínea (un gráfico de línea para cada producto). Mostrar una leyenda en la esquina superior izquierda con los colores de cada producto.

```python
productos=df.iloc[:,1:7]
plt.plot(meses,productos,marker='o',markersize=3, label=productos.columns, linewidth=2)
plt.xlabel('Numero de mes')
plt.ylabel('Unidades')
plt.legend(loc='upper left',prop={'size':7})
plt.xticks(meses)
plt.title('Unidades vendidas mensuales')
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000]);
```

### 8.- Leer las ventas de pasta dentífrica de cada mes y mostrarlas usando un gráfico de dispersión (*scatter plot*). Además, se mostrará el gráfico con  rejilla (*grid*). 

```python
pasta_dentrifica=df['crema_dentifrica']
plt.scatter(meses,pasta_dentrifica,label='Pasta dentrifica')
plt.grid(True,linestyle='--',linewidth=1)
plt.xticks(meses)
plt.title('Venta de dentrificos')
plt.legend(loc='upper left');
```

### 9.- Leer las ventas de crema facial y de crema limpiadora facial y mostrarlas en un gráfico de barras. El gráfico debe mostrar el número de unidades vendidas de cada producto por mes. Se debe mostrar una barra diferente para cada uno de esos dos productos.

```python
crema_facial=df['crema_facial']
crema_limpiadora=df['limpiador_facial']
plt.bar(x=list(map(lambda x: x - 0.125, meses)),height=crema_facial,width=0.25,label='Crema facial')
plt.bar(x=list(map(lambda x: x + 0.125, meses)),height=crema_limpiadora,width=0.25,label='Crema limpiadora facial')
plt.xticks(meses)
plt.grid(True,linestyle='--')
plt.legend(loc='upper left',prop={'size':7})

```

### 10.-	Leer las ventas de gel de baño de todos los meses y mostrarlo con un gráfico de barras. Guardar el gráfico en fichero (en Google Drive).

```python
gel_bano=df['gel_baño']
plt.bar(x=meses,height=gel_bano,label='Gel de bano')
plt.xticks(meses)
plt.grid(True,linestyle='--')
plt.legend(loc='upper left',prop={'size':7});



```

### 11.- Leer las ventas totales de cada mes y mostrarlo usando un histograma. Los intervalos del histograma (*bins*) serán de 25000 € y el rango completo va de 150000 300000.

```python
rangoVentas = list(range(150000, 300000, 25000))
plt.hist(beneficio_total,rangoVentas,rwidth=0.9)
print(beneficio_total)
```

### 12.- Calcular las ventas anuales totales de cada producto y mostrarlo en un gráfico de tarta. Mostrar en cada sector el porcentaje que representa su producto sobre el total.

```python
total_producto=productos.iloc[:,:].sum()
plt.figure(figsize=(9, 7))
plt.pie(total_producto,autopct='%1.1f%%',labels=total_producto.index)
plt.legend(loc='lower right');
plt.title('Ventas por producto');


```

### 13.- Leer las ventas de todos los meses de gel de baño y de limpiador facial y mostrarlos como dos subgráficos (subplots).

```python
plt.subplot(3,1,1)
plt.plot(meses,gel_bano)
plt.title('Ventas gel de bano')
plt.xticks(meses)
plt.subplot(3,1,3)
plt.plot(meses,crema_limpiadora)
plt.title('Ventas de limpiador facial')
plt.xticks(meses);
plt.xlabel('Numero de mes')
plt.ylabel('Unidades vendidas')
```
